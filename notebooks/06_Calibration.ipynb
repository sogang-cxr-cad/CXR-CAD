## 🌡️ 06. Model Calibration (확률값 보정)

모델이 80%라고 예측했을 때 실제 80%의 환자가 질환을 갖고 있는지, Confidence 값의 신뢰성을 맞추기 위해 교정합니다.

### 🎯 주요 분석 내용
- **Reliability Diagram 작성**: 예측 확률 구간별 실제 양성률을 비교하고, Overconfidence/Underconfidence 경향성 파악
- **Temperature Scaling 적용**: 로짓(Logit)에 Temperature 파라미터를 나누어주어 확률 분포 스케일링 전/후의 Expected Calibration Error(ECE) 수치 변화 계산
