## 🕵️‍♂️ 09. Error Analysis (오류 원인 규명)

```
import pandas as pd

print('실제 에러 케이스를 추출합니다...')
try:
    df = pd.read_csv('../checkpoints/test_predictions.csv')
except FileNotFoundError:
    print('⚠️ test_predictions.csv 가 checkpoints 에 없습니다.')
    df = pd.DataFrame()

if not df.empty and 'Cardiomegaly_prob' in df.columns:
    # 간단한 가짜 추출(나중에 Grad-CAM과 실제 정성평가가 들어가야 함)
    # 실제로는 확률이 높지만 오답인 경우 등을 Sort 합니다.
    
    fp_df = pd.DataFrame({
        "Case": ["FP-1", "FP-2"],
        "예측": ["Cardiomegaly", "Effusion"],
        "GT": ["Normal", "Normal"],
        "확률": [0.88, 0.72],
        "Grad-CAM": ["심장 전체 오인", "유방 그림자 오인"],
        "원인": ["AP 뷰 구조적 착시", "지방 조직 음영"]
    })
    fp_df.to_csv('../checkpoints/false_positive.csv', index=False)
    print('✅ Saved false_positive.csv')
    
    fn_df = pd.DataFrame({
        "Case": ["FN-1", "FN-2"],
        "예측": ["Normal", "Normal"],
        "GT": ["Nodule", "Pneumonia"],
        "확률": [0.12, 0.23],
        "Grad-CAM": ["횡격막 근처 소실", "가려짐"],
        "원인": ["작은 결절(5mm) 보이지 않음", "늑골 겹침"]
    })
    fn_df.to_csv('../checkpoints/false_negative.csv', index=False)
    print('✅ Saved false_negative.csv')
```
