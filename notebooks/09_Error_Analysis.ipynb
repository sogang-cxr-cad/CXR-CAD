## 🕵️‍♂️ 09. Error Analysis (Grad-CAM 등 오류 원인 규명)

왜 맞았는지 혹은 왜 틀렸는지 Black-Box 모델의 판단 근거를 시각화하고, False Positive(과잉 진단) 및 False Negative(미진단) 케이스의 분포를 뜯어봅니다.

### 🎯 주요 분석 내용
- **Grad-CAM 히트맵 추출 및 정성 분석**: 활성화 피처 맵이 병변을 정확히 가리키는지, 아니면 배경/기기(Pacemaker, 파이프)를 보고 컨닝(Shortcut Learning)했는지 검증
- **오답 샘플(False Case) Top N 수집**: 신뢰도가 높은 오답 케이스를 선별하여 임상적 시각에서 어떤 이유로 착각했는지 패턴 정리
